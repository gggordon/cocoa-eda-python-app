```python
from sklearn.model_selection import train_test_split
```

```python
X = data2[['Company(Makerifknown)','CompanyLocation','SpecificBeanOriginorBarName']]
y = data2['CocoaPercent']
```

```python
X.head()
```

```python
def create_label_encoder_dict(df):
    from sklearn.preprocessing import LabelEncoder
    
    label_encoder_dict = {}
    for column in df.columns:
        # Only create encoder for categorical data types
        if not np.issubdtype(df[column].dtype, np.number) or not np.issubdtype(df[column].dtype, np.float) :
            label_encoder_dict[column]= LabelEncoder().fit(df[column])
    return label_encoder_dict
```

```python
label_encoders = create_label_encoder_dict(X)
print("Encoded Values for each Label")
print("="*32)
for column in label_encoders:
    print("="*32)
    print('Encoder(%s) = %s' % (column, label_encoders[column].classes_ ))
    print(pd.DataFrame([range(0,len(label_encoders[column].classes_))], columns=label_encoders[column].classes_, index=['Encoded Values']  ).T)
```

```python

for column in X.columns:
    if column in label_encoders:
        X[column] = label_encoders[column].transform(X[column])

print("Transformed data set")
print("="*32)
X
```

```python
label_encoders
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
clf = DecisionTreeRegressor(max_depth=2) # Change the max_depth to 10 or another number
```

```python
0
```

```python
clf.fit(X_train, y_train)
```

```python
pd.DataFrame([ "%.2f%%" % perc for perc in (clf.feature_importances_ * 100) ], index = X_train.columns, columns = ['Feature Significance in Decision Tree'])
```
